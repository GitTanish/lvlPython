```python
import os

def create_text_file(filename="data.txt", num_rows=5, num_cols=3):
    """
    Creates a text file with a specified number of rows and columns,
    where each entry is a placeholder string like "Row1_Col1".  Handles
    potential errors during file operations.

    Args:
        filename (str, optional): The name of the text file to create.
            Defaults to "data.txt".
        num_rows (int, optional): The number of rows in the file.
            Defaults to 5.
        num_cols (int, optional): The number of columns in the file.
            Defaults to 3.
    """
    try:
        with open(filename, 'w') as file:
            for i in range(1, num_rows + 1):
                row_data = []
                for j in range(1, num_cols + 1):
                    row_data.append(f"Row{i}_Col{j}")
                # Join the column data with a comma or space
                file.write(','.join(row_data) + '\n')  # Use ',' for CSV, ' ' for space
        print(f"File '{filename}' created successfully in: {os.path.abspath(filename)}")

    except Exception as e:
        print(f"An error occurred while creating the file: {e}")
        return  # Important: Exit the function on error

def read_text_file(filename="data.txt"):
    """
    Reads and prints the contents of a text file.  Handles file not found
    and other potential errors.

    Args:
        filename (str, optional): The name of the text file to read.
            Defaults to "data.txt".
    """
    try:
        with open(filename, 'r') as file:
            content = file.read()
            print(f"\nContent of '{filename}':\n{content}")
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    except Exception as e:
        print(f"An error occurred while reading the file: {e}")

if __name__ == "__main__":
    # Create a text file
    create_text_file()

    # Read the text file
    read_text_file()

    # Create a text file with custom parameters
    create_text_file("my_data.txt", 10, 4)
    read_text_file("my_data.txt")
```
