```python
# agg method
# passing dict
genres.agg(
    {
        'Runtime':['mean','min'],
        'IMDB_Rating':['mean','sum'],
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# passing list
genres.agg({col: ['min', 'max', 'mean'] for col in movies.select_dtypes(include=['number']).columns})
# now in newer version of pandas we have to explicitly define where we are applying these agg functions
```

```python
# looping on groups
df = pd.DataFrame(columns=movies.columns)
top_movies = []
print(df)
for group, data in genres:
  # print(data[data['IMDB_Rating']==data['IMDB_Rating'].max()])
  # df = df.append(data[data['IMDB_Rating']==data['IMDB_Rating'].max()]) # this piece of code won't work because the
  top_movies.append(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])

df = pd.concat(top_movies, ignore_index=True)  # Concatenate all selected rows at once
print(df)                                                                        # append is no longer used in pandas 1.5 so,
```

```python
# split (apply) combine
# apply -> builtin function

genres.apply(min)
```

```python
# apply is best when you use them with custom functions
# find number of movies starting with A for each group

# def foo(group):
#   print(group)
#   return group

def foo(group):
  return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(foo)
```

```python
# find ranking of each movie in the group according to IMDB score
def rank_movie(group):
  group['genre_rank']=group['IMDB_Rating'].rank(ascending=False)
  return group
```

```python
genres.apply(rank_movie)
```

```python
# find normalized IMDB rating group wise
def normal(group):
  group['norm_rating'] = (group['IMDB_Rating']-group['IMDB_Rating'].min())/(group['IMDB_Rating'].max()-group['IMDB_Rating'].min())
  return group

genres.apply(normal)
```

```python
# groupby on multiple cols
duo=movies.groupby(['Director','Star1'])
# size
duo.size()
# get_group
duo.get_group(('Aamir Khan','Amole Gupte'))
```

```python
# find the most earning actor->director combo
duo['Gross'].sum().sort_values(ascending=False)
```

```python
# find the best(in-terms of metascore(avg)) actor->genre combo
# series when applied .reset_index() converts into dataframe
movies.groupby(['Star1','Genre'])['Metascore'].mean().reset_index().sort_values('Metascore', ascending=False)
```

```python
ipl= pd.read_csv('/content/deliveries.csv')
ipl.head()
```

```python
ipl.shape
```

```python
# find the top 10 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max no of sixes
six = ipl[ipl['batsman_runs'] == 6]

six.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(2)
```

```python
# find batsman with most number of 4's and 6's in last 5 overs
temp_df = ipl[ipl['over']>15]
temp_df[(temp_df['batsman_runs'] == 4) | (temp_df['batsman_runs'] == 4)]
temp_df.groupby('batsman')['batsman'].count().sort_values(ascending=False)
```

```python
# find V Kohli's record against all teams
temp_df = ipl[ipl['batsman'] == 'V Kohli']

temp_df.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
# Create a function that can return the highest score of any batsman
temp_df = ipl[ipl['batsman'] == 'V Kohli']
temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
def highest(batsman):
  temp_df = ipl[ipl['batsman'] == batsman]
  return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
highest('DA Warner')
```
