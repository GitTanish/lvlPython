### 2D Line plot

image.png

- Bivariate Analysis
- categorical -> numerical and numerical -> numerical
- Use case - Time series data

```python
# plotting a simple function
price   = [48000,54000,57000,49000,47000,45000] #numerical
year = [2015,2016,2017,2018,2019,2020] #categorical

# plt.plot(x,y)
plt.plot(year,price)
```

```python
# from a pandas dataframe
batsman=pd.read_csv('/content/sharma-kohli.csv')
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# label title
plt.plot(batsman['index'],batsman['V Kohli'],color= 'green')
plt.plot(batsman['index'],batsman['RG Sharma'], color='Red')

plt.title('Rohit Sharma vs Virat Kohli Carrer Comparison')
plt.xlabel('Season')
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color= 'green',linestyle='dashed',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'], color='Red',linestyle='dotted',linewidth=3)

plt.title('Rohit Sharma vs Virat Kohli Carrer Comparison')
plt.xlabel('Season')
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color= 'green',linestyle='solid',linewidth=3,marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'], color='Red',linestyle='solid',linewidth=3,marker='o',markersize=10)

plt.title('Rohit Sharma vs Virat Kohli Carrer Comparison')
plt.xlabel('Season')
plt.ylabel("Runs Scored")
```

```python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color= 'green',linestyle='solid',linewidth=3,marker='D',label="Virat")
plt.plot(batsman['index'],batsman['RG Sharma'], color='Red',linestyle='solid',linewidth=3,marker='o',markersize=10,label='Rohit')

plt.title('Rohit Sharma vs Virat Kohli Carrer Comparison')
plt.xlabel('Season')
plt.ylabel("Runs Scored")
plt.legend(loc='best')
```

```python
# limiting axes
price   = [48000,54000,57000,49000,47000,45000,450000] #numerical
year = [2015,2016,2017,2018,2019,2020,2021] #categorical

# since 450000 is an outlier the graph may show flat for most of the data
plt.plot(year,price,marker='o')
# trimming
plt.ylim(0,75000)
plt.xlim(2017,2019)
```

```python
# grid

plt.plot(batsman['index'],batsman['V Kohli'],color= 'green',linestyle='solid',linewidth=3,marker='D',label="Virat")
plt.plot(batsman['index'],batsman['RG Sharma'], color='Red',linestyle='solid',linewidth=3,marker='o',markersize=10,label='Rohit')

plt.title('Rohit Sharma vs Virat Kohli Carrer Comparison')
plt.xlabel('Season')
plt.ylabel("Runs Scored")
plt.legend(loc='best')
plt.grid()
```

```python
# show
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color= 'green',linestyle='solid',linewidth=3,marker='D',label="Virat")
plt.plot(batsman['index'],batsman['RG Sharma'], color='Red',linestyle='solid',linewidth=3,marker='o',markersize=10,label='Rohit')

plt.title('Rohit Sharma vs Virat Kohli Carrer Comparison')
plt.xlabel('Season')
plt.ylabel("Runs Scored")
plt.legend(loc='best')
plt.grid()
plt.show() # when the plotting is done in text editor
```

### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,30,60,50]
color = ['red','blue','green','yellow','pink']
plt.bar(color,children,color="Green")
```

```python
# horizontal bar chart
plt.barh(color,children,color='blue')
```

```python
# color and label
df = pd.read_csv("/content/batsman_season_record.csv")
df
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df["2015"],width=0.2)
plt.bar(np.arange(df.shape[0]),df["2016"],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df["2017"],width=0.2)

plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.show()
```

```python
np.arange(df.shape[0])
```

```python
# Problem
children = [10,20,30,60,50]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pink pink pink']
plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')
```

```python
# Stacked Bar chart

plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

### Pie Chart

image.png
- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
 # simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
# dataset
df = pd.read_csv('/content/gayle-175.csv')
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode,shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],labeldistance=1,shadow=True)
plt.show()
```
